LOAD PRETRAINED-MODEL FROM CHECKPOINT

```python
# Load pipeline config and build a detection model
configs = config_util.get_configs_from_pipeline_file(pipeline_path)
detection_model = model_builder.build(model_config=configs['model'], is_training=False)

# Restore checkpoint
ckpt = tf.compat.v2.train.Checkpoint(model=detection_model)
ckpt.restore(config_path).expect_partial()

@tf.function
def detect_fn(image):
    image, shapes = detection_model.preprocess(image)
    prediction_dict = detection_model.predict(image, shapes)
    detections = detection_model.postprocess(prediction_dict, shapes)
    return detections
```

filter_text function for filtering texts from licence plate

```python
def filter_text(roi , texts , t_h):

    t_h = (roi.shape[0] *roi.shape[1])*t_h
    plate = []
    for text in texts:
        length = np.sum(np.subtract(text[0][1], text[0][0]))
        height = np.sum(np.subtract(text[0][2], text[0][1]))

        if(length*height > t_h):
            plate.append(text)
    return plate

```

ocr_filter_visualize() function for extraction text from our region of interest using EASYOCR

```python
from collections import defaultdict

import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import easyocr
reader = easyocr.Reader(["en"])

def ocr_filter(image,detections,s_t , t_h):

     # Scores, boxes and classes above threhold
    scores = list(filter(lambda x: x> s_t, detections['detection_scores']))
    n = len(scores)
    boxes = detections['detection_boxes'][:n]
    classes = detections['detection_classes'][:n]

    #regoin of interest

    width = image.shape[1]
    height = image.shape[0]
    texts = defaultdict()
    for i , box in enumerate(boxes):

        roi = box*[height,width,height,width]

        region = image[int(roi[0]):int(roi[2]) , int(roi[1]):int(roi[3])]

        text = reader.readtext(region)
        plate = filter_text(region,text,t_h)
        for p in plate:
            texts[p[1]] = (box,scores[i])
    return texts
            
        
    
```

DETECTION FROM AN IMAGE

```python
img = cv2.imread(img_path)
image_np = np.array(img)

height = img.shape[0]
width = img.shape[1]

input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)
detections = detect_fn(input_tensor)

num_detections = int(detections.pop('num_detections'))
detections = {key: value[0, :num_detections].numpy()
              for key, value in detections.items()}
detections['num_detections'] = num_detections

# detection_classes should be ints.
detections['detection_classes'] = detections['detection_classes'].astype(np.int64)
texts = ocr_filter(image_np,detections,0.6,0.3)


for text,(roi,score) in texts.items():
        y1,x1,y2,x2 = roi*[height,width,height,width]
        cv2.rectangle(img,(int(x1),int(y1)) , (int(x2),int(y2)),(255,0,0),2)
        cv2.putText(img , f"{text} {score:.2}" , (int(x1) -40,int(y1)),cv2. FONT_HERSHEY_SIMPLEX, 1,(0,0,255),2,cv2.LINE_AA)
   
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

REAL TIME DETECTION FROM WEB CAMERA

```python
cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while cap.isOpened(): 
    ret, frame = cap.read()
    image_np = np.array(frame)
    
    input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0), dtype=tf.float32)
    detections = detect_fn(input_tensor)
    
    num_detections = int(detections.pop('num_detections'))
    detections = {key: value[0, :num_detections].numpy()
                  for key, value in detections.items()}
    detections['num_detections'] = num_detections

    # detection_classes should be ints.
    detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

    texts = ocr_filter(image_np,detections,0.8,0.3)
    for text,(roi,score) in texts.items():
        y1,x1,y2,x2 = roi*[height,width,height,width]
        cv2.rectangle(frame,(int(x1),int(y1)) , (int(x2),int(y2)),(255,0,0),2)
        cv2.putText(frame , f"{text} {score:.2}" , (int(x1),int(y1)),cv2. FONT_HERSHEY_SIMPLEX, 1,(255,0,0),2,cv2.LINE_AA)
   




    cv2.imshow("licence",frame)
    
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```
